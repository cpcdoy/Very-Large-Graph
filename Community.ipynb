```python
import sys, string, os, time, re
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

r = [f for f in os.listdir('louvain') if re.match(r'results.txt[0-9_]*', f)]

start_time = time.time()

print(r)

x = []
y = []
z = []
w = []
s = []

for res in r:
    res = "louvain/" + res
    f = open(res, 'r')
    data = f.read().splitlines()

    for i in range(len(data)):
        line = data[i].split()
        x.append(float(line[0]))
        y.append(float(line[1]))
        z.append(float(line[2]))
        w.append(float(line[3]))
        if (len(line) > 4):
            s.append(float(line[4]))
        else:
            s.append(float(line[0]) * (float(line[0]) - 1) * float(line[1]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='b', marker='o')

ax.set_xlabel('k')
ax.set_ylabel('K')
ax.set_zlabel('time')

plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, w, c='b', marker='o')
ax.set_zlabel('edge count')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, s, c='b', marker='o')
ax.set_zlabel('file size')
plt.show()
```
