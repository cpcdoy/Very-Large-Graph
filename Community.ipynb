```python
import sys, string, os, time
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

res = "louvain/results.txt"

start_time = time.time()
#a = os.system("./caveman2 100 100 " + res)

f = open(res, 'r')
data = f.read().splitlines()

print(len(data))
    
x = []
y = []
z = []
w = []


for i in range(len(data)):
    line = data[i].split()
    x.append(float(line[0]))
    y.append(float(line[1]))
    z.append(float(line[2]))
    w.append(float(line[3]))
```

```python
#1*1 to 100*100
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='b', marker='o', s=0.2)

ax.set_xlabel('k')
ax.set_ylabel('K')
ax.set_zlabel('time')

plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, w, c='b', marker='o', s=0.2)
ax.set_zlabel('nodes count')
plt.show()
```
